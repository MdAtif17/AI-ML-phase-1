# Mini Project

```python
# write a python program to create a calculator
print("=== Operator-Based Calculator ===")
print("Choose an operation:")
print(" +  : Addition")
print(" -  : Subtraction")
print(" * : Multiplication")
print(" /  : True Division")
print(" // : Floor Division (integer output)")
print(" %  : Modulo (remainder)")
print(" ** : Exponentiation (power)")
num1 = float(input("\nEnter first number: "))
operator = input("Enter an operator (+, -, *, /, //, %, **): ").strip()
num2 = float(input("Enter second number: "))
print("\n--- Result ---")
if operator == "+":
    result = num1 + num2
    print(f"{num1} + {num2} = {result}")
elif operator == "-":
    result = num1 - num2
    print(f"{num1} - {num2} = {result}")
elif operator == "*":
    result = num1 * num2
    print(f"{num1} * {num2} = {result}")
elif operator == "/":
  if num2 != 0:
        result = num1 / num2
        print(f"{num1} / {num2} = {result}")
  else:
        print("Error: Cannot divide by zero!")
elif operator == "//":
  if num2 != 0:
        result = num1 // num2
        print(f"{num1} // {num2} = {result} (Floor Division)")
  else:
        print("Error: Cannot divide by zero!")
elif operator == "%":
    if num2 != 0:
        result = num1 % num2
        print(f"{num1} % {num2} = {result} (Remainder)")
    else:
        print("Error: Cannot divide by zero!")
elif operator == "**":
    result = num1 ** num2
    print(f"{num1} raised to the power of {num2} = {result}")
else:
    print("Invalid operator entered!")
```
